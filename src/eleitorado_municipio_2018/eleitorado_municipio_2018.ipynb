# Análise do eleitorado brasileiro

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from jupyterthemes import jtplot

# choose which theme to inherit plotting style from
# onedork | grade3 | oceans16 | chesterish | monokai | solarizedl | solarizedd
jtplot.style(theme='onedork')

%matplotlib inline
```

```python
df = pd.read_csv('eleitorado_municipio_2018.csv', encoding='latin1', sep=';')

df[df['NM_MUNICIPIO'] == 'FORTALEZA'].T
```

```python
total_eleitores = df['QTD_ELEITORES'].sum()
total_masculino = df['QTD_ELEITORES_MASCULINO'].sum()
total_feminino = df['QTD_ELEITORES_FEMININO'].sum()
total_naoinformado = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
print('Total de leitores: ', total_eleitores)
print('Eleitores masc:', (total_masculino / total_eleitores * 100).round(2))
print('Eleitores fem: ', (total_feminino / total_eleitores * 100).round(2))
print('Eleitores Não informado: ', (total_naoinformado / total_eleitores * 100).round(2))
```

```python
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
df['RELACAO_MASC_FEM'] = df['QTD_ELEITORES_MASCULINO'] / df['QTD_ELEITORES_FEMININO']
```

```python
df_brasil = df[df['CD_PAIS'] == 1].copy()
```

```python
# Locais com mais eleitores homens que mulheres
df_brasil.nlargest(5, 'RELACAO_MASC_FEM').T
```

```python
# Locais com mais eleitoras mulheres que homens
df_brasil.nsmallest(5, 'RELACAO_MASC_FEM').T
```

```python
jtplot.style(theme='grade3')

df_brasil['RELACAO_MASC_FEM'].plot.hist(bins=100, figsize=(12,10))
```

```python
sns.distplot(df_brasil['RELACAO_MASC_FEM'], bins=100, color='red', kde=True)

plt.title('Relação eleitores Masc/Fem', fontsize=18)
plt.xlabel('Masc/Fem', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.axvline(1.0, color='black', linestyle='--')
```

```python
sns.swarmplot(data=df_brasil, x='NM_REGIAO', y='RELACAO_MASC_FEM')

plt.axhline(1.0, color='black', linestyle='--')
```
