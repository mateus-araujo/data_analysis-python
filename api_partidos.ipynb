## Análise dos dados da API de partidos
Análise dos dados obitidos pela API da Câmara dos Deputados

```python
import pandas as pd
import requests as request
```

```python
url = 'https://dadosabertos.camara.leg.br/api/v2/partidos?itens=100&ordem=ASC&ordenarPor=sigla'
response = request.get(url)
response.text
```

```python
# Converte response da API para dataframe

df = pd.DataFrame(response.json()['dados'])

df.head(40)
```

```python
df['nome'].count()
```

```python
# Retorna valor a partir do índice da coluna [linha, coluna]
df.iloc[10, 2]
```

```python
# Retorna valor a partir do nome da coluna [linha, coluna]
df.loc[10, 'sigla']
```
