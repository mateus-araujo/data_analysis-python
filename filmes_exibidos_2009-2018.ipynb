## Decisão sobre dados incorretos/faltantes/divergentes

```python
# df.shape retorne (linhas, colunas)

df.shape
```

```python
# Cria novo Dataframe sem relançamentos

df_new = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
# Mostra tipos da planilha

df_new.dtypes
```

```python
# Converte coluna para tipo data

df_new['Data de lançamento'] = pd.to_datetime(df_new['Data de lançamento'])
```

```python
df_new.dtypes
```

```python
df_new.head()
```

```python
# Quais anos tiveram mais lançamentos?

df_new['Data de lançamento'].dt.year.value_counts()
```

```python
# Plota o gráfico com lançamentos por ano
ax = df_new['Data de lançamento'].dt.year.value_counts(ascending=True).plot.barh(figsize=(10, 15), width=0.75)

# Insere labels de quantidade no gráfico
for i in ax.patches:
    ax.text(i.get_width()+.1, i.get_y()+.31, str(i.get_width()), fontsize=12)
```

```python
# Filme com maior bilheteria

df_new[df_new['Renda (R$) no ano de exibição'] == df_new['Renda (R$) no ano de exibição'].max()]
```

```python
# Filmes com maior bilheteria
# nlargest(qtd, column)

df_new.nlargest(10, 'Renda (R$) no ano de exibição')
```

```python
# Filmes com menor bilheteria

df_new[df_new['Renda (R$) no ano de exibição'] == df_new['Renda (R$) no ano de exibição'].min()]
```
